```python
import pandas as pd
import matplotlib.pyplot as plt
from pmdarima.arima import auto_arima
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np
```

```python
name = 'dataset_r3_2xlarge_us-east-1b.csv'
```

```python
df = pd.read_csv(name)
df
```

```python
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
# Converting 'DateTime' to datetime and extracting date
df['Date'] = pd.to_datetime(df['Time'])
df['Time'] = df['Date'].dt.date

# Grouping by 'Date' and calculating average price
average_price_per_day = df.groupby('Time')['Price'].mean().reset_index()

average_price_per_day
```

```python
average_price_per_day.to_csv(f"regular_{name}")
```
