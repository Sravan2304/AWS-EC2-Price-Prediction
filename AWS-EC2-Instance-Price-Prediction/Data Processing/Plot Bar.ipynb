```python
import matplotlib.pyplot as plt
import pandas as pd

# Data extracted from the provided image
models = ['ARIMA', 'SARIMA', 'LSTM', 'HOLT-WINTERS', 'PROPHET']
rmse_values = [0.0032363314181028126, 0.0008253516102654094, 0.0016565254035332743, 0.0015669540550866253, 0.0012091128260812451]
r2_values = [-0.5694485808638217, 0.8979250634046473, 0.5888149522409407, 0.6320797952661217, 0.5120900282633443]

# Creating a dataframe
df = pd.DataFrame({
    'Model': models,
    'RMSE': rmse_values,
    'R^2': r2_values
})

# Plotting RMSE values
plt.figure(figsize=(10, 6))

# RMSE bar plot
plt.subplot(1, 2, 1)
plt.bar(df['Model'], df['RMSE'], color='blue')
plt.title('RMSE Values')
plt.xlabel('Model')
plt.ylabel('RMSE')
plt.xticks(rotation=45, ha="right")

# R^2 bar plot
plt.subplot(1, 2, 2)
plt.bar(df['Model'], df['R^2'], color='green')
plt.title('R^2 Values')
plt.xlabel('Model')
plt.ylabel('R^2')
plt.xticks(rotation=45, ha="right")

# Adjust layout to prevent overlap
plt.tight_layout()

plt.savefig("m4_2x_large_ap_southeast_2a_no_neg.png")
# Display the plot
plt.show()
```
